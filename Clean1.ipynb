```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv('investments.csv')
```

```python
df
```

```python
#Software
boo = df['company_category_list'].str.contains('web|software|game|mobile|e-commerce|internet|advertising|ad targeting|education|social media|messag|security|apps|music|search|saas|travel|transportation', flags=re.IGNORECASE, na=False)
df['company_category_list'].mask(boo, 'Software', inplace=True)

#Business
boo = df['company_category_list'].str.contains('analytic|account|financ|consulting', flags=re.IGNORECASE, na=False)
df['company_category_list'].mask(boo, 'Business', inplace=True)

#Biotech
boo = df['company_category_list'].str.contains('bio', flags=re.IGNORECASE, na=False)
df['company_category_list'].mask(boo, 'Biotech', inplace=True)

#Industrial Tech
boo = df['company_category_list'].str.contains('3D|advanced materials|aerospace|clean technology|semiconductors|manufacturing|^technology$|nanotech|computers|information tech', flags=re.IGNORECASE, na=False)
df['company_category_list'].mask(boo, 'Industrial Tech', inplace=True)

#Health Care
boo = df['company_category_list'].str.contains('health|lifestyle|medical', flags=re.IGNORECASE, na=False)
df['company_category_list'].mask(boo, 'Health Care', inplace=True)

#Retail
boo = df['company_category_list'].str.contains('retail|fashion', flags=re.IGNORECASE, na=False)
df['company_category_list'].mask(boo, 'Retail', inplace=True)

#df.groupby('company_category_list')['company_category_list'].count().sort_values(ascending=False)
```

```python
#Drop Everything Else
boo = df['company_category_list'].isin(['Software', 'Business', 'Biotech', 'Industrial Tech', 'Health Care', 'Retail'])
df = df[boo]
df.groupby('company_category_list')['company_category_list'].count().sort_values(ascending=False)
```

```python
#Save
df.to_csv('investments_clean.csv')
```
